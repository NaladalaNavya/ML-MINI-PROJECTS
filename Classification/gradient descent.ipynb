<h1 color="blue">Gradient Descent</h1>

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
def gradient_descent(x,y):
    m_current=b_current=0
    iterations=10000
    n=len(x)
    learning_rate = 0.01
    for i in range(iterations):
        y_predicted=m_current*x +b_current
        plt.plot(x,y_predicted,color="green")
        cost = (1/n)*sum((val**2 for val in (y-y_predicted)))
        mderivative = (-2/n)*sum(x*(y-y_predicted))
        bderivative = (-2/n)*sum(y-y_predicted)
        m_current = m_current-learning_rate*mderivative
        b_current = b_current-learning_rate * bderivative
        #print("m= {},b= {},cost= {},iteration={}".format(m_current,b_current,cost,i))
     
```

```python
x=np.array([1,2,3,4])
y=np.array([5,6,7,9])
```

```python
gradient_descent(x,y)
```
